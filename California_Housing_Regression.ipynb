# Support Vector Regressor

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# 1. Initialize and Fit the Model
# 'C' controls the penalty for points outside the tube; 'epsilon' is the tube width
svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
svr_model.fit(X_train, y_train)

# 2. Make Predictions
svr_predictions = svr_model.predict(X_test)

# 3. Calculate Metrics
svr_mse = mean_squared_error(y_test, svr_predictions)
svr_r2 = r2_score(y_test, svr_predictions)

print(f"SVR MSE: {svr_mse:.4f}")
print(f"SVR R2 Score: {svr_r2:.4f}")
```

Support Vector Regression (SVR) works on the principle of finding a hyperplane that best fits the data while staying within a specific threshold ($\epsilon$). It uses the Kernel Trick to transform the data into a higher-dimensional space, making it possible to find a linear fit for non-linear data.
